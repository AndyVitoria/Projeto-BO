```python
grid_size = (28,31)

def new_grid():
    grid = []
    for i in range(grid_size[0]):
        #grid[i] = []
        grid.append([])
        for j in range(grid_size[1]):
            #grid[i][j] = 0
            grid[i].append(0)
    return grid
```

```python
import pandas as pd

df = pd.read_excel('dataframe_6.xlsx')
df
```

```python
grid = new_grid()

for reg in df.values:
    col = reg[3]
    row = reg[10]
    grid[col][row] += 1
```

```python
import seaborn as sns;

sns.heatmap(grid)
```
